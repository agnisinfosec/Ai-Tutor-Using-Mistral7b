#Text File


```python
import requests

url = "https://raw.githubusercontent.com/hwchase17/chat-your-data/master/state_of_the_union.txt"
res = requests.get(url)
with open("state_of_the_union.txt", "w") as f:
  f.write(res.text)
```

```python
pip install -U langchain-community
```

```python
# Document Loader
from langchain.document_loaders import TextLoader
loader = TextLoader('./state_of_the_union.txt')
documents = loader.load()
```

```python
documents[:100]
```

```python
import textwrap

def wrap_text_preserve_newlines(text, width=110):
    # Split the input text into lines based on newline characters
    lines = text.split('\n')

    # Wrap each line individually
    wrapped_lines = [textwrap.fill(line, width=width) for line in lines]

    # Join the wrapped lines back together using newline characters
    wrapped_text = '\n'.join(wrapped_lines)

    return wrapped_text
```

```python
print(wrap_text_preserve_newlines(str(documents[0])))
```

```python
# Text Splitter
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)
```

```python
len(docs)
```

```python
docs[3]
```

# Embeddings


```python
# Embeddings
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings()
```

```python
# Vectorstore: https://python.langchain.com/en/latest/modules/indexes/vectorstores.html
from langchain.vectorstores import FAISS

db = FAISS.from_documents(docs, embeddings)
```

```python
query = "What did the president say about the Supreme Court tell in one line"
docs = db.similarity_search(query)
```

```python
print(wrap_text_preserve_newlines(str(docs[0].page_content)))
```

#Working with PDFs Files


```python
!pip install PyPDF2 -U -q

import PyPDF2
from langchain.document_loaders import UnstructuredPDFLoader
from langchain.indexes import VectorstoreIndexCreator
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import CharacterTextSplitter
import os
```

```python
# connect your Google Drive
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)

pdf_folder_path = '/content/sample_data'
os.listdir(pdf_folder_path)
```

```python
def preprocess_pdf(pdf_path):
    """Attempts to fix structural issues in the PDF."""
    try:
        with open(pdf_path, 'rb') as f:
            reader = PyPDF2.PdfReader(f)
            writer = PyPDF2.PdfWriter()

            for page in reader.pages:
                writer.add_page(page)

            with open(pdf_path, 'wb') as outfile:
                writer.write(outfile)

    except Exception as e:
        print(f"Error preprocessing {pdf_path}: {e}")
```

```python
# Preprocess each PDF file
for fn in os.listdir(pdf_folder_path):
    if fn.endswith(".pdf"):
        pdf_path = os.path.join(pdf_folder_path, fn)
        preprocess_pdf(pdf_path)


loaders = [UnstructuredPDFLoader(os.path.join(pdf_folder_path, fn)) for fn in os.listdir(pdf_folder_path) if fn.endswith(".pdf")]

index = VectorstoreIndexCreator(
    embedding=HuggingFaceEmbeddings(),
    text_splitter=CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)).from_loaders(loaders)
```

```python
llm=HuggingFaceEndpoint(repo_id="mistralai/Mistral-7B-Instruct-v0.2", temperature=0.1, max_length=512)
```

```python
from langchain.chains import RetrievalQA
chain = RetrievalQA.from_chain_type(llm=llm,
                                    chain_type="stuff",
                                    retriever=index.vectorstore.as_retriever(),
                                    input_key="question")
```

```python
chain.run('what is self attention? explain in short and crisp manner')
```

```python
chain.run('How was the Mistral model trained?')
```
